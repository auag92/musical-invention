```python
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.linear_model import Ridge
from sklearn.linear_model import LinearRegression
import matplotlib.pyplot as plt
```

```python
def ms_error(y1, y2):
    return np.sqrt(np.mean((y1-y2)**2))

def ridge_fit(x, y, lmda):
    return np.matmul(np.linalg.inv(np.matmul(x.T, x) + 
                                   lmda * np.eye(x.shape[1])), 
                     np.matmul(x.T, y))

def ridge_predict(x, w):
    w = np.squeeze(w)[:,None]
    return np.matmul(x, w)
```

```python
x_data = np.loadtxt("X.dat")
y_data = np.loadtxt("Y.dat")

print(x_data.shape, y_data.shape)

print(x_data.dtype)
```

```python
plt.hist(y_data, bins=20)
plt.title("distributing of target values")
plt.show()
```

```python
for i in range(x_data.shape[1]):
    print(np.corrcoef(x_data[:,i], y_data)[0,1])
```

```python
X_train, X_test, y_train, y_test = x_data[:80], x_data[80:], y_data[:80], y_data[80:] 
# X_train, X_test, y_train, y_test = train_test_split(x_data, y_data, test_size=0.20, random_state=42)
```

```python
model = Ridge(alpha=1.0)
model.fit(X_train, y_train)
y_hat = model.predict(X_train)
y_pred = model.predict(X_test)
```

```python
w = ridge_fit(X_train, y_train, lmda=1.0)
y_hat = ridge_predict(X_train, w)
y_pred = ridge_predict(X_test, w)
```

```python
print("training error", ms_error(y_hat, y_train))
print("test error", ms_error(y_pred, y_test))

plt.plot(y_hat, y_train, "r*")
plt.plot(y_pred, y_test, "g*")
plt.title("parity plot")
plt.legend(["training", "test"])
plt.show()
```

```python
print("training error", ms_error(y_hat, y_train))
print("test error", ms_error(y_pred, y_test))

plt.plot(y_hat, y_train, "r*")
plt.plot(y_pred, y_test, "g*")
plt.title("parity plot")
plt.legend(["training", "test"])
plt.show()
```
